```python


# Given list of product prices
prices = [450, 1200, 899, 1500, 300]
threshold = 1000
count = 0
expensive_products = []

for price in prices:
    if price > threshold:
        count += 1
        expensive_products.append(price)
        print(f"  ✓ Found expensive product: ₹{price}")

print(f"\nProduct prices: {prices}")
print(f"Products above {threshold}: {count}")
print(f"Those expensive products are: {expensive_products}")
expensive = [p for p in prices if p > threshold]
count_short = len(expensive)
print(f"Products above {threshold}: {count_short}")
print(f"List of expensive items: {expensive}")

# Method 3: Using filter function
expensive_filtered = list(filter(lambda x: x > threshold, prices))
print(f"Products above {threshold}: {len(expensive_filtered)}")

# Real-life example with product names
print("\n" + "=" * 60)
print("Real-Life Application: E-commerce Filter")
print("=" * 60)
products = {
    "Mobile Phone": 15000,
    "Headphones": 800,
    "Laptop": 45000,
    "Mouse": 500,
    "Tablet": 20000
}

print(f"Filter: Products above ₹{threshold}")
premium_products = {name: price for name, price in products.items() if price > threshold}

for name, price in premium_products.items():
    print(f"  • {name}: ₹{price}")

print(f"\nTotal premium products: {len(premium_products)}")
```
