```python

# Given dictionary
employees = {"Aditya": 75000, "Devyani": 68000, "Prajakta": 72000}
if "Aditya" in employees:
    print("Employee exists")

print("\n" + "=" * 60)
print("Detailed Examples")
print("=" * 60)

# Example 1: Check single employee
print("\nExample 1: Checking if 'Aditya' exists")
if "Ravi" in employees:
    print("✓ Employee exists")
    print(f"  Salary: ₹{employees['Aditya']}")
else:
    print("✗ Employee does not exist")
print("\nExample 2: Checking if 'Snehal' exists")
if "John" in employees:
    print("✓ Employee exists")
else:
    print("✗ Employee does not exist")

# Example 3: Check multiple employees
print("\nExample 3: Check multiple employees")
names_to_check = ["Aditya", "Snehal", "Devyani", "Prajakta"]

for name in names_to_check:
    if name in employees:
        print(f"✓ {name} exists")
    else:
        print(f"✗ {name} does not exist")

# Example 4: Different methods to check
print("\n" + "=" * 60)
print("Different Methods to Check Key")
print("=" * 60)
if "Aditya" in employees:
    print("Employee exists")
print("\nMethod 2: Using keys() method")
if "Devyani" in employees.keys():
    print("Employee exists")
print("\nMethod 3: Using get() method")
result = employees.get("Kiran")
if result is not None:
    print("Employee exists")

# Real-Time Application
print("\n" + "=" * 60)
print("Real-Time Application: User Login System")
print("=" * 60)

users = {
    "Admin": "admin123",
    "Aditya": "pass123",
    "Devyani": "secure456"
}

# Simulate login
username = "Aditya"
print(f"\nLogin attempt for: {username}")

if username in users:
    print("✓ User exists - Proceed to password verification")
else:
    print("✗ User not found - Please register")

# Another example
username2 = "Snehal"
print(f"\nLogin attempt for: {username2}")

if username2 in users:
    print("✓ User exists")
else:
    print("✗ User not found - Please register")

print("\n" + "=" * 60)
print("Configuration Verification Example")
print("=" * 60)

config = {
    "database": "localhost",
    "port": 5432,
    "username": "admin"
}

# Check if required keys exist
required_keys = ["database", "port", "username", "password"]

print("\nChecking configuration...")
all_present = True

for key in required_keys:
    if key in config:
        print(f"✓ {key}: Present")
    else:
        print(f"✗ {key}: Missing")
        all_present = False

if all_present:
    print("\n✓ Configuration is complete!")
else:
    print("\n✗ Configuration is incomplete!")
```
