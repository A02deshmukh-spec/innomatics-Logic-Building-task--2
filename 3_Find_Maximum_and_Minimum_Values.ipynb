```python

# Given list of numbers
numbers = [45, 22, 89, 10, 66]
maximum = max(numbers)
minimum = min(numbers)

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
max_value = numbers[0]  # Start by assuming first number is the biggest
min_value = numbers[0]  # Start by assuming first number is the smallest

print(f"Starting with first number: {numbers[0]}")

for num in numbers:
        if num > max_value:
            max_value = num  # Found a bigger number!
            print(f"  Found new maximum: {num}")
        if num < min_value:
            min_value = num  # Found a smaller number!
            print(f"  Found new minimum: {num}")

print(f"\nFinal Result:")
print(f"Max: {max_value}")
print(f"Min: {min_value}")

# Real-life example
print("\n" + "=" * 60)
print("Real-Life Application:")
print("=" * 60)
temperatures = [25, 18, 32, 15, 28]
print(f"Weekly temperatures (°C): {temperatures}")
print(f"Hottest day: {max(temperatures)}°C")
print(f"Coldest day: {min(temperatures)}°C")
print(f"Temperature range: {max(temperatures) - min(temperatures)}°C")
```
